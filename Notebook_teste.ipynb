```python
#Leitura de dados TITANIC

import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 

dataSource = pd.DataFrame(pd.read_csv("test.csv"))
```

```python
#Graficos de homens em comparação a mulheres 


genderMale_count = 0
genderFemale_count = 0

for people in dataSource['Sex']:
    if people == 'male':
        genderMale_count+=1
    else:
        genderFemale_count+=1

colors = ['#6A5ACD','#483D8B']

x= np.array(['Homens', 'Mulheres'])
y = np.array([genderMale_count, genderFemale_count])
plt.bar(x,y, color = colors)
plt.show
```

```python

passangersClasses = [0,0,0]

for classes in dataSource['Pclass']:
   if classes == 3: 
       passangersClasses[2]+=1
   elif classes == 2:
       passangersClasses[1]+=1
   elif classes == 1:
        passangersClasses[0]+=1
   else:
       break

classesNames = ['1st', '2nd','3rd']
colors_classes = ['#A9A9A9','#C0C0C0','#1C1C1C']


plt.bar(classesNames, passangersClasses, color = colors_classes)
plt.show
```
